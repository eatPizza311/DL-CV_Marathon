# 知識點
- 齊次座標(Homogenious coordinate) 簡單來說就是將 n 維的座標用 n+1 維的方式表示。
    - 二維座標：$(x, y)$
    - 二維齊次座標：$(x', y', w')$
    - 二維齊次座標通式：$(kx', ky', kw')$
    
  這些表達方式都代表同一個點，而座標間的關係為 $x=\frac{x'}{w'}$、$y=\frac{y'}{w'}$。
  
  
- 齊次座標多出來的一個維度可以解釋成「遠近」，因此當 $w'=0$ 就可以代表無限遠的點。
    <img src='img/homo.png' width=250>
  
- 各種轉換其實都可以理解成「圖片上的點用齊次座標表示後乘上一個轉換矩陣 M」

    $$
    \begin{bmatrix}
    a& b & c\\ 
    d& e & f\\ 
    g& h &i
    \end{bmatrix}
    \begin{bmatrix}
    u\\ 
    v\\ 
    w\end{bmatrix} = 
    \begin{bmatrix}
    x'\\ 
    y'\\ 
    w'\end{bmatrix}
    $$
    
    其中 $a、b、d、e$ 負責線性轉換，$c、f$ 負責平移，$g、h$ 負責透視，三維範例請參考 [齊次座標](https://openhome.cc/Gossip/ComputerGraphics/HomogeneousCoordinate.htm)


- 在 OpenCV 中的實作中只要給予轉換前後的4個點座標，point1 = …，point2 = …，然後用`M = cv2.getPerspectiveTransform(point1, point2)` 取得變換矩陣，接著再用透視變換的 function 轉換 (與`cv2.warpAffine()`類似)
```python
img_new = cv2.warpPerspective(img, M, (col, row))
```

# 範例

根據以下的參考點，嘗試做透視變換

```
point1 = np.array([[60, 40], [420, 40], [420, 510], [60, 510]], dtype=np.float32)
point2 = np.array([[0, 80], [w, 120], [w, 430], [0, 470]], dtype=np.float32)
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('img/lena.png')
````

## 透視轉換

````python
img_perspective = img.copy()
h, w = img.shape[:2]

# 設定四對點，並取得 perspective 矩陣
point1 = np.array([[60, 40], [420, 40], [420, 510], [60, 510]], dtype=np.float32)
point2 = np.array([[0, 80], [w, 120], [w, 430], [0, 470]], dtype=np.float32)
M = cv2.getPerspectiveTransform(point1, point2)

# perspective 轉換
img_perspective = cv2.warpPerspective(img, M, (w, h))

# 在圖片上標記點
img_copy = img.copy()
for idx, pts in enumerate(point1):
    pts = tuple(map(int, pts))
    cv2.circle(img_copy, pts, 5, (0, 255, 0), -1)
    cv2.putText(img_copy, str(idx), (pts[0]+5, pts[1]+5), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)

for idx, pts in enumerate(point2):
    pts = tuple(map(int, pts))
    cv2.circle(img_perspective, pts, 5, (0, 255, 0), -1)
    cv2.putText(img_perspective, str(idx), (pts[0]+5, pts[1]+5), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)


# 組合 + 顯示圖片
img_show = np.hstack((img_copy, img_perspective))
img_show = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 5))
plt.imshow(img_show)
plt.axis('off');
````

1. 思考一下如果經過透視變換，是否會因為視角轉換的差異導致部份 pixel 遺失造成無法修復的問題？
2. 透視變換是否有甚麼實際上的應用？

````python
# 設定四對點，並取得 perspective 矩陣
point1 = np.array([[0, 80], [w, 120], [w, 430], [0, 470]], dtype=np.float32)
point2 = np.array([[0, 0], [w, 0], [w, h], [0, h]], dtype=np.float32)
M = cv2.getPerspectiveTransform(point1, point2)
img_perspective_2 = cv2.warpPerspective(img_perspective, M, (w, h))

plt.figure(figsize=(12, 5))
img_show = np.hstack((img_copy, img_perspective_2))
img_show = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show)
plt.axis('off');
````
