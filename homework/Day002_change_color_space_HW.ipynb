# 知識點
1. 除了常見的 RGB 以外，還有以下常見色彩表示法:
    - HSB、HSV - 色相(Hue)、飽和度(Saturation)、明度(Brightness、Value)
        <img src='hsv.png' width=200>
        其中 Hue 以 360 度決定顏色<br>
        Saturation 以 0~100% 決定色彩純度，白、黑、灰色系色彩的飽和度為 0<br>
        Brightness 以 0~100% 決定鮮豔程度，黑色明度為 0<br><br>
        
    - HSL - 色相(Hue)、飽和度(Saturation)、亮度(Lightness)
        <img src='hsl.png' width=200>
        
     HSL 在飽和度跟明亮度與 HSB 的表示有點不同
     
     **飽和度**
        - HSB 的飽和度 0~100% 是白色到選擇的色相。
        - HSL 的飽和度 0~100% 是灰色到選擇的色相，他們認為純白色是飽和的顏色。
        
     **亮度**
        - HSB 的明度是光的量，可以是任何顏色，換句話說，B=100 時看 H, S 決定是甚麼顏色。
        - HSL 的亮度是白色的量，換句話說，L=100 時是白色。
        
     比較兩者可以發現 HSL 的白色在最上面，HSB/HSV 的白色在圓心。
     
    - Lab
        <img src='lab.png' width=200>
        
        與 RGB 相比，可以表示的顏色範圍更大也更接近人眼可以感知的範圍。<br>
        其中 L = Lighness 以 0 ~ 100 決定明亮度，數值由小到大，由黑到白。<br>
        A 以 -128 ~ 127 代表顏色對立的維度，數值由小到大，由綠到紅。<br>
        B 以 -128 ~ 127 代表顏色對立的維度，數值由小到大，由藍到黃。
2. 使用 `cv2.cvtColor(IMAGE, CODE)` 改變圖片的呈現。

   其中 IMAGE 為要轉換的圖片，CODE 為色彩空間轉換的目標，例如
   ```python
   cv2.cvtColor('lena.png', cv2.COLOR_BGR2HSV) # Convert BGR to HSV
   ```
   更多轉換碼請參考 [ColorConversionCodes](https://docs.opencv.org/3.2.0/d7/d1b/group__imgproc__misc.html#ga4e0972be5de079fed4e3a10e24ef5ef0)

````python
import cv2
img_path = 'lena.png'
# 以彩色圖片的方式載入
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
# 改變不同的 color space
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
````

````python
# 為了要不斷顯示圖片，所以使用一個迴圈
while True:
    cv2.imshow('bgr', img)
    cv2.imshow('hsv', img_hsv)

    # 直到按下 ESC 鍵才會自動關閉視窗結束程式
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
````

````python
import matplotlib.pyplot as plt
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(121);plt.imshow(img_rgb)
plt.xticks([]), plt.yticks([])  
plt.subplot(122);plt.imshow(img_hsv) 
plt.xticks([]), plt.yticks([])  
plt.show()
````

- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

````python
img_hsl = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
````

````python
# 為了要不斷顯示圖片，所以使用一個迴圈
while True:
    cv2.imshow('bgr', img)
    cv2.imshow('hsv', img_hsv)
    cv2.imshow('hsl', img_hsl)
    cv2.imshow('lab', img_lab)

    # 直到按下 ESC 鍵才會自動關閉視窗結束程式
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
````

````python
plt.figure(figsize=(20, 16))
plt.subplot(141);plt.imshow(img_rgb)
plt.xticks([]), plt.yticks([])  
plt.subplot(142);plt.imshow(img_hsv) 
plt.xticks([]), plt.yticks([])
plt.subplot(143);plt.imshow(img_hsl)
plt.xticks([]), plt.yticks([])
plt.subplot(144);plt.imshow(img_lab)
plt.xticks([]), plt.yticks([])
plt.show()
````
