# 知識點
## 透過 color space 轉換改圖
Data Augmentation 有時候也會透過 color space 的轉換去更改我們想要改變的顏色屬性 (飽和度、明亮度...)。

例如想要調整明亮度，會考慮把圖片轉到 HSV (或是其他有明亮度的 color space) 再去調整明亮度的值。

而調整特定顏色屬性牽涉到指定要操作的 channel，例如想要改變明亮度就必須使用 `img[:, :, 2]` 或 `img[..., 2]` 選取明度軸。

在我們選定要操作的範圍後還需要特別注意

OpenCV 會以 0~255 的值表示
- 有些值是以百分比或是包含負值的方式表示

OpenCV 顯示圖片會以 uint8 (可儲存範圍 0~255) 方式表示
- 如果計算過程包含小數或是負值要自己轉換

實做範例如下：
```python
# 轉換color space (BGR —> HSV)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# 操作飽和度需要轉為百分比 (浮點數)
img_hsv = img_hsv.astype('float32')
# 轉成浮點數後，把數值範圍縮放到[0, 1]，調降兩成飽和度
img_hsv[..., 1] = img_hsv[..., 1]/255 - 0.2

# 後面再做一些邊緣條件判斷，確保所有值都在合理範圍內 (0~1)
# 把圖片依序轉回 0~255 區間 > uint8格式 > RGB 就可以顯示了
```
### 改變飽和度

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# read image as color format
img = cv2.imread('lena.png', cv2.IMREAD_COLOR)

# 先把 color space 轉為 HSL 格式 (OpenCV 表示順序是 HLS)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
change_percent = 0.2
sat_ax = 2

# 針對飽和度的值做改變，超過範圍 0~1 的都會截斷
# 在 HLS color space 減少飽和度
img_hls_down = img_hls.astype('float32')
img_hls_down[..., sat_ax] = img_hls_down[..., sat_ax]/255. - change_percent
img_hls_down[img_hls_down[..., sat_ax] < 0] = 0 # 截斷小於 0 的值
img_hls_down[..., sat_ax] = img_hls_down[..., sat_ax]*255
img_hls_down = img_hls_down.astype('uint8')
img_hls_down = cv2.cvtColor(img_hls_down, cv2.COLOR_HLS2RGB)

# 在 HLS color space 增加飽和度
img_hls_up = img_hls.astype('float32')
img_hls_up[..., sat_ax] = img_hls_up[..., sat_ax]/255. + change_percent
img_hls_up[img_hls_up[..., sat_ax] > 1] = 1 # 截斷大於 1 的值
img_hls_up[..., sat_ax] = img_hls_up[..., sat_ax]*255
img_hls_up = img_hls_up.astype('uint8')
img_hls_up = cv2.cvtColor(img_hls_up, cv2.COLOR_HLS2RGB)

# 為了在 matplotlib 正確顯示
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 組合 + 顯示圖片
img_hls_change = np.hstack((img_rgb, img_hls_down, img_hls_up))
plt.imshow(img_hls_change);
````

## 直方圖均衡化 (Histogram Equalization)
當圖片對比不強時(顏色的值集中在某個區域)，可以透過直方圖均衡化來增強對比(透過數學公式將值分散到 0~255 區間)。

這種方法對於背景和前景都太亮或者太暗的圖像非常有用，這種方法尤其是可以帶來X光圖像中更好的骨骼結構顯示以及曝光過度或者曝光不足照片中更好的細節。

其 OpenCV 實作為 `cv2.equalizeHist(img)`
<img src='hist_equalization.png' width=500>

詳細說明請參考 [灰階直方圖均化 Histogram Equalization](https://jason-chen-1992.weebly.com/home/-histogram-equalization)
> 雖然直方圖均衡主要是處理灰圖，如果處理 RGB 圖通常是會轉到 HSV 再對明亮度做直方圖均衡，不過我們也可以個別對 RGB 的 3 個 channel 做直方圖均衡。




````python
# 轉為灰階圖片
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 灰階圖片直方圖均衡
img_gray_equal = cv2.equalizeHist(img_gray)

# 組合圖片 + 顯示圖片
img_gray_equalHist = np.hstack((img_gray, img_gray_equal))
plt.imshow(img_gray_equalHist, cmap='gray');
````

## 調整對比度 / 明亮度
直方圖均衡是參考全圖做對比度跟明亮度的調整，但我們也可以簡單透過乘法跟加法來手動調整。

$$g(x) = \alpha f(x) + \beta$$

其中
- $g(x)$ 為調整過的像素值。
- $f(x)$ 為原始像素值。
- $\alpha$ 調整對比度 ($\alpha$>0)。
- $\beta$ 調整明亮度。
> 因為像素值的範圍為 [0, 255]，所以參數值不能設定太大，官方建議為 $\alpha \in (1.0, 3.0)$、$\beta \in (0, 100)$。

實作可以透過 for 迴圈逐一調整像素值
```python
for b in range(img.shape[0]):
    for g in range(img.shape[1]):
        for r in range(img.shape[2]):
            # 透過 np.clip 把範圍限制在 0~255
            img[b, g, r] = np.clip(alpha*img[b, g, r]+beta, 0, 255)
```
或直接透過 OpenCV 的內建函式實作 `img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)`

````python
# alpha: 控制對比度 (1.0~3.0)
# beta: 控制明亮度 (0~255)
add_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=0)
add_lighness = cv2.convertScaleAbs(img, alpha=1.0, beta=50)

# 組合圖片 + 顯示圖片
img_contrast_light = np.hstack((img, add_contrast, add_lighness))
img_contrast_light = cv2.cvtColor(img_contrast_light, cv2.COLOR_BGR2RGB)
plt.imshow(img_contrast_light);
````

2. 實作直方圖均衡

## 改變飽和度

- 轉換成 HSV color space, 改變 s channel 的值

````python
# 為了要改變飽和度，我們先把 color space 轉成 HSV 格式
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
change_percentage = 0.2

# 針對飽和度的值做改變，超過界線 0~1 的都會 bound
# 在 HSV color space 減少飽和度
img_hsv_down = img_hsv.astype('float32')
img_hsv_down[..., 1] = img_hsv_down[..., 1]/255. - change_percentage
img_hsv_down[img_hsv_down[..., 1] < 0] = 0
img_hsv_down[..., 1] = img_hsv_down[..., 1]*255
img_hsv_down = img_hsv_down.astype('uint8')

# 在 HSV color space 增加飽和度
img_hsv_up = img_hsv.astype('float32')
img_hsv_up[..., 1] = img_hsv_up[..., 1]/255. + change_percentage
img_hsv_up[img_hsv_up[..., 1] > 1] = 1
img_hsv_up[..., 1] = img_hsv_up[..., 1]*255
img_hsv_up = img_hsv_up.astype('uint8')


# 轉換 color space 回 RGB
img_hsv_down = cv2.cvtColor(img_hsv_down, cv2.COLOR_HSV2RGB)
img_hsv_up = cv2.cvtColor(img_hsv_up, cv2.COLOR_HSV2RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 組合圖片 + 顯示圖片
img_hsv_change = np.hstack((img_rgb, img_hsv_down, img_hsv_up))
plt.figure(figsize=(15, 12))
plt.xticks([]);plt.yticks([])
plt.imshow(img_hsv_change);
````

## 直方圖均衡

1. case 1: 把彩圖拆開對每個 channel 個別做直方圖均衡再組合起來
2. case 2: 轉換 color space 到 HSV 之後對其中一個 channel 做直方圖均衡

````python
# case 1
# 每個 channel 個別做直方圖均衡
equalHist_by_channel_r = cv2.equalizeHist(img[:, :, 2])
equalHist_by_channel_g = cv2.equalizeHist(img[:, :, 1])
equalHist_by_channel_b = cv2.equalizeHist(img[:, :, 0])

# 組合經過直方圖均衡的每個 channel
img_bgr_equal = cv2.merge([equalHist_by_channel_r, equalHist_by_channel_g, equalHist_by_channel_b])

# case 2 - 轉換 color space 後只對其中一個 channel 做直方圖均衡
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsv[..., 2] = cv2.equalizeHist(img_hsv[..., 2])
img_hsv_equal = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

# # 組合圖片 + 顯示圖片
img_bgr_equalHist = np.hstack((img_rgb, img_bgr_equal, img_hsv_equal))
plt.figure(figsize=(15, 12))
plt.xticks([]);plt.yticks([])
plt.imshow(img_bgr_equalHist);
````

## 調整對比 / 明亮

````python
# alpha: 控制對比度 (1.0~3.0)
# beta: 控制明亮度 (0~255)
add_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=0)
add_lighness = cv2.convertScaleAbs(img, alpha=1.0, beta=50)

# 組合圖片 + 顯示圖片
img_contrast_light = np.hstack((img, add_contrast, add_lighness))
img_contrast_light = cv2.cvtColor(img_contrast_light, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 12))
plt.xticks([]);plt.yticks([])
plt.imshow(img_contrast_light);
````
