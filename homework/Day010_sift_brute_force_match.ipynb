# 知識點
取得特徵之後，就代表我們已經有能力去識別關鍵點的特殊性，在這之後可以接到許多電腦視覺的任務：
- 配對：判斷兩張圖片上相同物體的位置
- 辨識：判斷兩張圖片上是否有相同物體
- 全景圖：尋找兩張圖片的相同視角，再經過轉換合成全景圖 [Feature Matching & Automatically Stitching 360 Panoramas](http://www.cs.cmu.edu/afs/andrew/scs/cs/15-463/f05/pub/www/projects/fproj/mrom/)

廣泛的說，SIFT 只是其中一種抽取特徵的方式，這邊會延續上一章節以 SIFT 為例介紹配對的應用。

## SIFT 配對
<img src='img/sift_match.png' width=700>

配對會計算兩張圖片的關鍵點在特徵空間上的距離，若小於某閾值就視為是相同的特徵，在 SIFT 特徵的配對任務中，通常會使用 L2 norm 的方式計算：

$$d(\mathbf{p, q}) = d(\mathbf{q, p}) = \sqrt{(q_1-p_1)^2+(q_2-p_2)^2+...+(q_n-p_n)^2}$$

兩個 128 維的向量根據上面公式計算可以得到一個距離。

簡單暴力的配對方法是把 query image 的關鍵點逐一針對每個 train image 的關鍵點計算 L2 距離：
- 取得距離最小的配對
- 取得 k 個最適合的配對

為了確保配對的合適性，可以先在計算時取得 k 個配對，再根據距離過濾不適合的配對。

例如尋找 k=2 個最好的配對方式，透過 ratio test 的方式來過濾一些不適當的配對，因為有時候 query 的關鍵點並不會出現在 train image 中，參考原論文的 [Section 7.1](https://www.cs.ubc.ca/~lowe/papers/ijcv04.pdf)，有提到配對任務中 ratio test 目的以及建議比值設定在 0.7~0.8 比較好。

## SIFT 與機器學習

許多機器學習的任務一開始都要先經過抽取特徵的步驟，SIFT 等傳統電腦視覺的特徵只是其中一種方式，深度學習則是另外一種抽取特徵的方式。

後面再根據任務類型選擇要對特徵做甚麼處理，所以傳統特徵跟 model 抽的特徵，在用途上是差不多的。

例如分類任務，抽取出來的的特徵可以當作 input 輸入分類器。

SIFT 雖然可以做機器學習任務，但是實作上存在一些問題，因為演算法不保證所有圖片都會產生一樣維度的特徵：
- SIFT 每一個特徵點的維度其實是一樣的，但每張圖片產生的特徵點個數不同，才會導致圖片的特徵維度不同。
- 其中一種作法是做 Clustering，每一張圖片都取 n 個特徵點來固定圖片的特徵維度。
- 缺點在於當圖片太簡單導致部份圖片特徵太少就會失效，所以類似 MNIST 或是 CIFAR 等簡單的資料集就不太適合。

# 範例

透過 SIFT 特徵實作 Brute-Force Matching

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 以灰階方式讀入圖片
img_query = cv2.imread('img/box.png', 0)
img_train = cv2.imread('img/box_in_scene.png', 0)

# 建立 SIFT 物件
sift = cv2.xfeatures2d_SIFT.create()

# 偵測並計算 SIFT 特徵 (keypoints 關鍵點, descriptor 128 維敘述子)
kp_query, des_query = sift.detectAndCompute(img_query, None)
kp_train, des_train = sift.detectAndCompute(img_train, None)
````

## 基於 SIFT 特徵的暴力比對

- D.Lowe ratio test
- knn 比對

````python
# 建立 Brute-Force Matching 物件
bf = cv2.BFMatcher(cv2.NORM_L2)

# 以 knn 方式暴力比對特徵
matches = bf.knnMatch(des_query, des_train, k=2)

# 透過 D.Lowe ratio test 排除不適合的配對
candidate = []
for m, n in matches:
    if m.distance < 0.75*n.distance:
        candidate.append([m])

# 顯示配對結果
img_show = cv2.drawMatchesKnn(img_query, kp_query, img_train, kp_train, candidate, None, flags=2)

plt.figure(figsize=(20, 18))
plt.axis('off')
plt.imshow(img_show, cmap='gray')
````
